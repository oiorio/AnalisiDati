## Python e IPython

- IPython  un interpete "interattivo" alternativo a quello standard di python da linea di comando.

- Python  concepito come un linguaggio interpretato, in realt nella sua versione pi comune preoduce dei files .pyc che sono delle "caches" in bytecode compilate, che vengono riutilizzate all'occorrenza girando il python file.

Nota: La distinzione tra interpretato e compilato  da molti considerata "old fashioned", perch molti linguaggi presentano un'implementazione compilata, interpretata, o both.

```python
h1 = ROOT.TH1F("h1_esempio","Istogramma d'esempio",100,-10,10)
f1 = ROOT.TF1("f1_esempio","gaus",-10,10)
f1.SetParameters(10,1,5)

h1.FillRandom("f1_esempio",1000)

c1 = ROOT.TCanvas()
c1.Draw()

h1.Draw()
```

```python
c1.Draw()

for i in range(0,200):
    h1.FillRandom("f1_esempio",1)
h1.DrawNormalized()
h1.Fit(f1,"S")
```
